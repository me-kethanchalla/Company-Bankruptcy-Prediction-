```python
import pandas as pd
import numpy as np

def group_and_drop_by_target_correlation(df, target, corr_threshold=0.95):
    

    # List all feature columns (excluding target)
    features = [col for col in df.columns if col != target]
    
    # Compute the absolute correlation matrix among the features
    corr_matrix = df[features].corr().abs()
    
    # Compute absolute correlation of each feature with the target
    target_corr = df[features].corrwith(df[target]).abs()
    
    # Group features that are highly correlated with each other.
    groups = []  # list to hold groups (each group is a list of feature names)
    visited = set()  # to track features already placed in a group
    
    for i, col in enumerate(features):
        if col in visited:
            continue
        # Start a new group with the current feature
        group = [col]
        for other in features[i+1:]:
            # If correlation between col and other is above the threshold, add other to the group
            if corr_matrix.loc[col, other] > corr_threshold:
                group.append(other)
                visited.add(other)
        groups.append(group)
        visited.add(col)
    
    # Now, for each group, select the feature that has the highest correlation with the target
    features_to_keep = []
    groups_selected = {}  # for logging: chosen representative mapped to the full group
    
    for group in groups:
        if len(group) == 1:
            # No redundancy in this group, keep the feature
            features_to_keep.append(group[0])
        else:
            # For a group of correlated features, pick the one with highest target correlation.
            best_feature = max(group, key=lambda x: target_corr[x])
            features_to_keep.append(best_feature)
            groups_selected[best_feature] = group
    
    # Create a new DataFrame with only the target and the selected features.
    df_reduced = df[[target] + features_to_keep]
    
    return df_reduced, groups_selected

def correlation_order_with_target(df, target):
   
   
    features = [col for col in df.columns if col != target]
    target_corr = df[features].corrwith(df[target]).abs().sort_values(ascending=False)
    return target_corr

# ---------------------------
# Example Usage:
# ---------------------------
target = 'Bankrupt?'

# Group and drop redundant features based on correlation threshold and target correlation
df_reduced, groups_selected = group_and_drop_by_target_correlation(df, target, corr_threshold=0.95)

# Print out the groups and the representative selected from each group
print("Highly correlated feature groups and their selected representative:")
for rep, group in groups_selected.items():
    print(f"Selected '{rep}' from group {group}")
```

```python

target_corr_order = correlation_order_with_target(df_reduced, target)
print("\nFeatures sorted by absolute correlation with the target:")
print(target_corr_order.to_string())
```

```python
# ----- Step 1: Imports (reuse previously imported libraries as needed) -----
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
)
from sklearn.utils.class_weight import compute_class_weight
from scikeras.wrappers import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# ----- Step 2: Updated Evaluation Function -----
def evaluate_model(model, X_train, y_train, X_test, y_test, model_name="Model", fit_params=None):
    if fit_params is None:
        fit_params = {}
    model.fit(X_train, y_train, **fit_params)
    y_pred = model.predict(X_test)
    
    # Try to get predicted probabilities (for ROC-AUC)
    try:
        y_proba = model.predict_proba(X_test)[:, 1]
    except Exception:
        y_proba = None

    accuracy = accuracy_score(y_test, y_pred)
    precision_0 = precision_score(y_test, y_pred, pos_label=0)
    precision_1 = precision_score(y_test, y_pred, pos_label=1)
    recall_0 = recall_score(y_test, y_pred, pos_label=0)
    recall_1 = recall_score(y_test, y_pred, pos_label=1)
    f1_0 = f1_score(y_test, y_pred, pos_label=0)
    f1_1 = f1_score(y_test, y_pred, pos_label=1)
    roc_auc = roc_auc_score(y_test, y_proba) if y_proba is not None else None

    print("\n" + "=" * 60)
    print(f"📌 **{model_name} Performance Metrics**")
    print("=" * 60)
    print(f"🔹 **Accuracy**          : {accuracy:.4f}")
    if roc_auc is not None:
        print(f"🔹 **ROC-AUC Score**     : {roc_auc:.4f}")
    print("-" * 60)
    print(f"🔹 **Precision (Class 0 - Not Bankrupt)** : {precision_0:.4f}")
    print(f"🔹 **Precision (Class 1 - Bankrupt)**     : {precision_1:.4f}")
    print(f"🔹 **Recall (Class 0 - Not Bankrupt)**    : {recall_0:.4f}")
    print(f"🔹 **Recall (Class 1 - Bankrupt)**        : {recall_1:.4f}")
    print(f"🔹 **F1-Score (Class 0 - Not Bankrupt)**  : {f1_0:.4f}")
    print(f"🔹 **F1-Score (Class 1 - Bankrupt)**      : {f1_1:.4f}")
    print("=" * 60)
    print("\n🔍 **Confusion Matrix**")
    print(confusion_matrix(y_test, y_pred))
    print("=" * 60)

# ----- Step 3: Data Preparation (reuse your existing DataFrame 'df') -----
# Assuming your target column is named "Bankrupt?"
X = df.drop(columns=['Bankrupt?'])
y = df['Bankrupt?']

# Stratified split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# For neural networks, scale the features
scaler = StandardScaler()
X_train_nn = scaler.fit_transform(X_train)
X_test_nn = scaler.transform(X_test)

# ----- Step 4: Define the Neural Network Model Function -----
def create_nn_model(input_dim):
    model = Sequential()
    model.add(Dense(64, input_dim=input_dim, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))  # Binary classification
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# ----- Step 5: Create and Evaluate the Neural Network Without Class Weighting -----
nn_model = KerasClassifier(
    model=create_nn_model,
    model__input_dim=X_train_nn.shape[1],
    epochs=50,
    batch_size=32,
    verbose=0
)

evaluate_model(
    nn_model,
    X_train_nn,
    y_train,
    X_test_nn,
    y_test,
    model_name="Neural Network (No Weighting)"
)

# ----- Step 6: Create and Evaluate the Neural Network With Class Weighting -----
# Compute class weights using scikit-learn for target column "Bankrupt?"
classes = np.unique(y_train)
cw = compute_class_weight(class_weight="balanced", classes=classes, y=y_train)
class_weight_dict = dict(zip(classes, cw))

nn_model_weighted = KerasClassifier(
    model=create_nn_model,
    model__input_dim=X_train_nn.shape[1],
    epochs=50,
    batch_size=32,
    verbose=0
)

evaluate_model(
    nn_model_weighted,
    X_train_nn,
    y_train,
    X_test_nn,
    y_test,
    model_name="Neural Network (Class Weighted)",
    fit_params={"class_weight": class_weight_dict}
)
```
